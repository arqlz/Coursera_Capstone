## Adquiring zip codes from Santiago City

```python
address = 'Santiago, Republica Dominicana'
g = geocoder.google(address, key=GOOGLE_APIKEY )
loc_stgo = g.latlng
print(address, "location", loc_stgo)
```

```python
sgo_barrios = ["Altos de Rafey", "Altos de Virella", "La Arboleda", "El Área Monumental", "Arroyo Hondo Abajo", "Baracoa", "Bella Vista", "Buena Vista", "Buenos Aires", "Centro de la Ciudad", "Cienfuegos", "El Dorado", "Ensanche Bermúdez", "Ensanche Bolívar", "Ensanche Libertad", "Ensanche Espaillat", "Gregorio Luperón", "Gurabo", "Hato Mayor", "Hoya del Caimito", "Jardines del Este", "La Joya", "La Trinitaria", "La Yaguita de Pastor", "La Zurza", "Las Antillas", "Las Colinas", "Los Álamos,↵Los Cerros de Gurabo", "Los Ciruelitos", "Los Jardines Metropolitanos", "Los Jazmines", "Los Rieles", "Los Salados", "Los Santos", "Mejoramiento Social", "Monterico", "Nibaje", "Parque Metropolitano de Santiago", "Pekín", "Peralta", "Pontezuela al Medio", "Pueblo Nuevo", "Reparto Consuelo", "Reparto del Este", "Reparto Universitario", "Rincón Largo", "Urbanización Henríquez", "Urbanización Los Reyes", "Villa Olga", "Villa Olímpica", "Zona Industrial"]
sgo_barrios = sgo_barrios + ["Guarabo Arriba", "Guarabo Abajo", "Los Molinos", "Los Prados", "Los Salados", "El Embrujo II", "El Embrujo III", "Cerro Hermoso", "La Primavera", "Cerro Alto", "Bella Vista"]

r = []
index = 0
for barrio in sgo_barrios:
    clear_output(wait=True)
    index = index+1
    print("Buscando ubiacion de ", barrio, index, "de", len(sgo_barrios))
    g = geocoder.google("{}, Santiago, Republica Dominicana".format(barrio), key=google_api_key)  
    r.append([barrio, g.latlng[0], g.latlng[1]])
  
    

sgo = pd.DataFrame(r, columns=["Lugar", "Latitud", "Longitud"])
sgo
```

```python
sgo.to_csv("sgo_neighborhoods.csv", index=False)
```

```python
map = folium.Map(location=loc_stgo, zoom_start=12)
for index,row in sgo.iterrows():
    label = '{}'.format(row.Lugar)
    label = folium.Popup(label, parse_html=True)
    folium.Circle(
        [row["Latitud"], row["Longitud"]],  popup=label,  radius=650,  color="red",   fill=True,   fill_color="red",   fill_opacity=0.2,    parse_html=False).add_to(map)  

map
```
